```python
from itertools import product
import numpy as np
import pandas as pd
from model_settings import vanilla_pricer
vanilla = vanilla_pricer.vanilla_pricer()
s = 100
K = np.linspace(s*0.5,s*1.5,5).astype(int).astype(float).tolist()
T = np.arange(1,365+1,1)
r = [0.04]
vols = np.linspace(0.1,0.9,5)

features = pd.DataFrame(
    product(
        [s],
        K,
        T,
        r,
        vols,
        ['call','put']
    ),
    columns = ['spot_price','strike_price','days_to_maturity','risk_free_rate','volatility','w']
)
features['price'] = vanilla.df_numpy_black_scholes(features)
features
```
