```python
import pandas as pd
import numpy as np
```

```python
from historical_av_key_collector import h5_name, keys, dates
```

```python
keys_df = pd.DataFrame(
    np.empty((len(keys),4),dtype=object),
    columns = ['raw_data_key','surface_key','calibration_key','parameter_key']
)
keys_df['raw_data_key'] = keys
```

```python
for i,row in keys_df.iterrows():
    key = row['raw_data_key']
    locator = key[:key.find('/',1)+1]
    keys_df.at[i,'surface_key'] = str(locator+'surface')
    keys_df.at[i,'calibration_key'] = str(locator+'heston_calibration/calibration_results')
    keys_df.at[i,'parameter_key'] = str(locator+'heston_calibration/heston_parameters')
```

```python
keys_df
```

```python
chain = {}
with pd.HDFStore(h5_name) as store:
    for i,row in keys_df.iterrows():
        link = {}
        link['raw_data'] = store[row['raw_data_key']]
        link['surface'] = store[row['surface_key']]
        link['calibration'] = store[row['calibration_key']]
        link['parameters'] = store[row['parameter_key']]
        chain[i] = link
```
